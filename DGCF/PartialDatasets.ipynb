```python
import numpy as np
import pandas as pd
import shutil
```

```python
data = pd.read_csv("data/mooc.csv", header=0, index_col=False)
data
```

```python
rates = [0.1, 0.05, 0.01, 0.005]
percents = ["10", "5", "1", "0.5"]
for rate, percent in zip(rates, percents):
    for i in range(10):
        drop_indices = np.random.choice(data.index, int(len(data)*(1-rate)), replace=False)
        data1 = data.drop(drop_indices).reset_index(drop=True)
        data1.to_csv("data/mooc/mooc" + str(percent) + "%_" + str(i) + ".csv", index=False)
    data1 = data[:int(len(data) * rate)]
    data1.to_csv("data/mooc/mooc" + str(percent) + "%_10.csv", index=False)
```
